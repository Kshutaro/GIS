```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefiles for each layer
suigaiH1=gpd.read_file('02水害ハザードマップ/水害ハザードメッシュ-1m以上.zip')
suigaiH2=gpd.read_file('02水害ハザードマップ/水害ハザードメッシュ-20cm〜50cm.zip')
suigaiH3=gpd.read_file('02水害ハザードマップ/水害ハザードメッシュ-20cm未満.zip')
# Create a new figure and axis
fig, ax = plt.subplots(figsize=(10, 10))

# Plot each layer on the axis
suigaiH1.plot(ax=ax, color='purple', label='suigai1')
suigaiH2.plot(ax=ax, color='blue', label='suigai2')
suigaiH3.plot(ax=ax, color='lightblue', label='suigai3')

# Set the title and legend
ax.set_title('Geopandas Map Plot with Three Layers')
ax.legend()

# Show the plot
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt


# Create a new figure and axis
iryou=gpd.read_file('01防災マップ/医療拠点施設.zip')
hinan=gpd.read_file('01防災マップ/避難場所シンボル.zip')
hukushihinan=gpd.read_file('01防災マップ/福祉避難所.zip')
fig, ax = plt.subplots(figsize=(10, 10))

# Plot each layer on the axis
iryou.plot(ax=ax, color='red', label='医療施設')
hinan.plot(ax=ax, color='blue', label='避難場所')
hukushihinan.plot(ax=ax, color='green', label='福祉避難所')

# Set the title and legend
ax.set_title('Geopandas Map Plot with Three Layers')
ax.legend()

# Show the plot
plt.show()
```
